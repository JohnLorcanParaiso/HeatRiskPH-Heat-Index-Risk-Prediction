# 🔧 Phase 3 · Data Preparation

**CRISP-DM Stage 3 — Prepare the dataset for modeling**

This phase cleans the dataset, creates new features, and prepares the target variables for regression, classification, and time series forecasting.

## 3.1 Feature Engineering

```python
# Create a copy of the original dataset for cleaning and feature engineering --
df_clean = df.copy()

# Convert the date column into datetime format for extracting date-related features --
df_clean["date"] = pd.to_datetime(df_clean["date"])

# Extract from the date column --
df_clean["year"] = df_clean["date"].dt.year
df_clean["month"] = df_clean["date"].dt.month
df_clean["day"] = df_clean["date"].dt.day
df_clean["day_of_year"] = df_clean["date"].dt.dayofyear
df_clean["weekday"] = df_clean["date"].dt.weekday

# Define function to assign Philippine seasonal categories based on month --
def assign_ph_season(month):
    if month in [11, 12, 1, 2]:
        return "Cool-Dry"   # November to February
    elif month in [3, 4, 5]:
        return "Hot-Dry"    # March to May
    else:
        return "Rainy"      # June to October

# Apply seasonal classification to each row based on month --
df_clean["season"] = df_clean["month"].apply(assign_ph_season)

print("Feature engineering completed.")

# Display selected engineered features for preview --
display(df_clean[["date", "month", "day_of_year", "weekday", "season"]].head())
```

## 3.2 Column Cleaning and Renaming

```python
rename_dict = {
    "temperature_2m_mean (°C)": "temp_mean",
    "temperature_2m_max (°C)": "temp_max",
    "temperature_2m_min (°C)": "temp_min",
    "apparent_temperature_mean (°C)": "heat_index_mean",
    "apparent_temperature_max (°C)": "heat_index_max",
    "apparent_temperature_min (°C)": "heat_index_min"
}

df_clean = df_clean.rename(columns=rename_dict)

print("Columns cleaned and renamed successfully.")
print(df_clean.columns.tolist())

display(df_clean.head())
```

## 3.3 Missing Value Treatment

```python
print("Missing values before treatment:")
display(df_clean.isnull().sum())

numeric_cols = df_clean.select_dtypes(include=["number"]).columns
df_clean[numeric_cols] = df_clean[numeric_cols].interpolate(method="linear")

df_clean = df_clean.ffill().bfill()

print("Missing values after treatment:")
display(df_clean.isnull().sum())
```

## 3.4 Outlier Checking

```python
numeric_cols = df_clean.select_dtypes(include=["number"]).columns

# Exclude date-related numeric columns from outlier checking --
columns_to_exclude = ['year', 'month', 'day', 'day_of_year', 'weekday']
numeric_cols = [col for col in numeric_cols if col not in columns_to_exclude]

for col in numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df_clean[col])
    plt.title(f"Boxplot of {col}")
    plt.xlabel(col)
    plt.show()

print("Outlier checking completed for selected numeric columns.")
```

## 3.5 Heat Risk Category Creation

```python
# Define function to classify heat risk levels based on heat index value --
def classify_heat_risk(value):
    if value < 27:
        return "Safe"
    elif value < 33:
        return "Caution"
    elif value < 42:
        return "Extreme Caution"
    elif value < 52:
        return "Danger"
    else:
        return "Extreme Danger"

# Apply heat risk classification to mean heat index values --
df_clean["heat_risk"] = df_clean["heat_index_mean"].apply(classify_heat_risk)

print("Heat risk categories created successfully.")

display(df_clean[["date", "heat_index_mean", "heat_risk"]].head())

print("\nHeat Risk Distribution:")
display(df_clean["heat_risk"].value_counts())
```

## 3.6 Final Modeling Dataset

```python
# Convert seasonal categories into dummy/one-hot encoded variables --
season_dummies = pd.get_dummies(df_clean["season"], prefix="season", dtype=int)

# Combine original dataset with generated seasonal dummy columns --
df_temp = pd.concat([df_clean, season_dummies], axis=1)

# Define expected seasonal columns to ensure consistency --
expected_season_cols = ["season_Cool-Dry", "season_Hot-Dry", "season_Rainy"]

# Add missing seasonal columns with default value 0 if not present --
for col in expected_season_cols:
    if col not in df_temp.columns:
        df_temp[col] = 0

# Select feature columns to be used for modeling --
feature_cols = [
    "temp_mean",
    "temp_max",
    "temp_min",
    "heat_index_max",
    "heat_index_min",
    "season_Cool-Dry",
    "season_Hot-Dry",
    "season_Rainy"
]

# Define important column names --
date_col = "date"
target_regression = "heat_index_mean"
target_classification = "heat_risk"

# Create final dataset containing date, features, and target variables --
model_df = df_temp[
    [date_col] + feature_cols + [target_regression, target_classification]
].copy()

print("Final modeling dataset created successfully.")

print("Shape:", model_df.shape)

display(model_df.head())
```

# 🧠 Phase 4 · Modeling

**CRISP-DM Stage 4 — Build predictive models**

This project uses three models:
1. Regression
2. Classification
3. Time Series Forecasting

## 4.1 Model 1 · Regression: Predicting Heat Index

```python
# Separate feature variables and regression target variable --
X_reg = model_df[feature_cols]
y_reg = model_df[target_regression]

# Split dataset into training and testing sets for regression --
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg,
    y_reg,
    test_size=0.20,
    random_state=42
)

# Initialize Random Forest Regressor model --
reg_model = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)

# Train regression model using training data --
reg_model.fit(X_train_reg, y_train_reg)

# Generate predictions on test data --
y_pred_reg = reg_model.predict(X_test_reg)

# Calculate regression evaluation metrics --
mae_reg = mean_absolute_error(y_test_reg, y_pred_reg)
rmse_reg = np.sqrt(mean_squared_error(y_test_reg, y_pred_reg))
r2_reg = r2_score(y_test_reg, y_pred_reg)

print("Regression Model Results")
print("MAE:", round(mae_reg, 3))
print("RMSE:", round(rmse_reg, 3))
print("R²:", round(r2_reg, 3))
```

## 4.2 Model 2 · Classification: Classifying Heat Risk Levels

```python
# Separate feature variables and classification target variable --
X_class = model_df[feature_cols]
y_class = model_df[target_classification]

# Count occurrences of each class label --
class_counts = y_class.value_counts()

# Check if all classes have at least 2 samples for stratified splitting --
can_stratify = class_counts.min() >= 2

# Split dataset into training and testing sets for classification --
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(
    X_class,
    y_class,
    test_size=0.20,
    random_state=42,
    stratify=y_class if can_stratify else None
)

# Initialize Random Forest Classifier model --
class_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight="balanced"
)

# Train classification model using training data --
class_model.fit(X_train_class, y_train_class)

# Generate predictions on test data --
y_pred_class = class_model.predict(X_test_class)

# Calculate classification accuracy --
accuracy_class = accuracy_score(y_test_class, y_pred_class)

print("Classification Model Results")
print("Accuracy:", round(accuracy_class, 3))

print("\nClassification Report:")
print(classification_report(y_test_class, y_pred_class))
```

```python
# Classification Visualization

risk_order = ["Safe", "Caution", "Extreme Caution", "Danger", "Extreme Danger"]
existing_labels = [label for label in risk_order if label in sorted(set(y_test_class) | set(y_pred_class))]

cm = confusion_matrix(y_test_class, y_pred_class, labels=existing_labels)
cm_df = pd.DataFrame(cm, index=existing_labels, columns=existing_labels)

plt.figure(figsize=(9, 7))
sns.heatmap(cm_df, annot=True, fmt="d", cmap="Oranges")
plt.title("Classification Model: Confusion Matrix")
plt.xlabel("Predicted Heat Risk")
plt.ylabel("Actual Heat Risk")
plt.show()

pred_risk_counts = pd.Series(y_pred_class).value_counts().reindex(existing_labels, fill_value=0)

plt.figure(figsize=(9, 5))
plt.bar(pred_risk_counts.index, pred_risk_counts.values)
plt.title("Predicted Heat Risk Level Distribution")
plt.xlabel("Heat Risk Level")
plt.ylabel("Number of Days")
plt.xticks(rotation=30)
plt.grid(axis="y", alpha=0.3)
plt.show()

display(pred_risk_counts)
```

## 4.3 Model 3 · Time Series Forecasting: Forecasting Heat Index

### 4.3.1 Prepare Time Series Dataset

```python
# Create a copy of the modeling dataset for time series analysis --
ts_df = model_df.copy()

# Convert date column into datetime format --
ts_df["date"] = pd.to_datetime(ts_df["date"])

# Sort dataset by date, set date as index, and enforce daily frequency --
ts_df = ts_df.sort_values("date").set_index("date")
ts_df = ts_df.asfreq("D")

# Define target variable for time series forecasting --
target_ts = "heat_index_mean"

# Define exogenous variables to be used in SARIMAX model --
sarimax_exog_cols = [
    "season_Cool-Dry",
    "season_Hot-Dry",
    "season_Rainy"
]

# Ensure all required exogenous columns exist --
for col in sarimax_exog_cols:
    if col not in ts_df.columns:
        ts_df[col] = 0

# Fill missing target values using time-based interpolation --
ts_df[target_ts] = ts_df[target_ts].interpolate(method="time")

# Forward-fill missing exogenous values, then replace remaining missing values with 0 --
ts_df[sarimax_exog_cols] = ts_df[sarimax_exog_cols].ffill().fillna(0)

# Extract target series as float type --
y_ts = ts_df[target_ts].astype(float)

# Extract exogenous variables as float type --
exog_ts = ts_df[sarimax_exog_cols].astype(float)

print("SARIMAX dataset prepared successfully.")
print("Date range:", y_ts.index.min(), "to", y_ts.index.max())
print("Total observations:", len(y_ts))
display(ts_df[[target_ts] + sarimax_exog_cols].head())
```

### 4.3.3 Test Stationarity

```python
# Define function to perform stationarity tests on a time series --
def stationarity_tests(series, series_name="Series"):
    # Remove missing values before testing --
    series = series.dropna()

    # Print header for stationarity test results --
    print(f"\nStationarity Tests for: {series_name}")
    print("-" * 60)

    # Perform Augmented Dickey-Fuller (ADF) test --
    adf_result = adfuller(series)
    print("ADF Test")
    print("ADF Statistic:", round(adf_result[0], 4))
    print("p-value:", round(adf_result[1], 4))

    # Interpret ADF test result --
    if adf_result[1] <= 0.05:
        print("ADF Interpretation: Stationary")
    else:
        print("ADF Interpretation: Non-stationary")

    print()

    # Perform KPSS test inside try-except block to handle possible errors --
    try:
        kpss_result = kpss(series, regression="c", nlags="auto")
        print("KPSS Test")
        print("KPSS Statistic:", round(kpss_result[0], 4))
        print("p-value:", round(kpss_result[1], 4))

        if kpss_result[1] > 0.05:
            print("KPSS Interpretation: Stationary")
        else:
            print("KPSS Interpretation: Non-stationary")

    except Exception as e:
        print("KPSS Test could not be completed:", e)

# Run stationarity tests on original heat index time series --
stationarity_tests(y_ts, "Original Heat Index Mean")

# Apply first differencing to remove trend/non-stationarity --
y_ts_diff = y_ts.diff().dropna()

# Run stationarity tests on differenced series --
stationarity_tests(y_ts_diff, "First-Differenced Heat Index Mean")
```

### 4.3.5 Model Selection Using AIC and BIC

```python
# Calculate 80% of the dataset length for training data --
train_size = int(len(y_ts) * 0.80)

# Split target time series into training and testing sets --
y_train_ts = y_ts.iloc[:train_size]
y_test_ts = y_ts.iloc[train_size:]

# Split exogenous variables into training and testing sets --
exog_train_ts = exog_ts.iloc[:train_size]
exog_test_ts = exog_ts.iloc[train_size:]

# Define parameter search ranges for SARIMAX hyperparameter tuning --
p_values = range(0, 2)
d_values = [1]
q_values = range(0, 2)

# Define seasonal SARIMAX parameter ranges --
P_values = range(0, 2)
D_values = range(0, 2)
Q_values = range(0, 2)

# Set weekly seasonality (7 days) --
seasonal_period = 7

# Store model evaluation results --
results = []

print("Starting optimized Grid Search... This may take a few minutes.")

# Loop through all parameter combinations --
for order in itertools.product(p_values, d_values, q_values):
    for seasonal_params in itertools.product(P_values, D_values, Q_values):

        # Build seasonal order tuple --
        seasonal_order = (
            seasonal_params[0],
            seasonal_params[1],
            seasonal_params[2],
            seasonal_period
        )

        try:
            # Initialize SARIMAX model with current parameters --
            model = SARIMAX(
                y_train_ts,
                exog=exog_train_ts,
                order=order,
                seasonal_order=seasonal_order,
                trend="n",
                enforce_stationarity=False,
                enforce_invertibility=False
            )

            # Fit model with limited iterations for faster execution --
            model_fit = model.fit(disp=False, maxiter=50)

            # Save model parameters and AIC score --
            results.append({
                "order": order,
                "seasonal_order": seasonal_order,
                "aic": model_fit.aic
            })

        # Skip invalid parameter combinations --
        except:
            continue

# Convert results into DataFrame --
results_df = pd.DataFrame(results)

# Check if grid search found valid models --
if len(results_df) == 0:
    best_order = (1, 1, 1)
    best_seasonal_order = (0, 0, 1, 7)
else:
    # Sort models by lowest AIC (best model) --
    results_df = results_df.sort_values("aic").reset_index(drop=True)

    # Display top 5 best parameter combinations --
    display(results_df.head(5))

    # Extract best model parameters --
    best_order = results_df.loc[0, "order"]
    best_seasonal_order = results_df.loc[0, "seasonal_order"]

# Print best SARIMAX parameters found --
print("Best SARIMAX Order:", best_order)
print("Best Seasonal Order:", best_seasonal_order)
```

### 4.3.6 Fit Final SARIMAX Model

```python
# Initialize SARIMAX model using the best parameters from grid search --
best_sarimax_model = SARIMAX(
    y_train_ts,
    exog=exog_train_ts,
    order=best_order,
    seasonal_order=best_seasonal_order,
    trend="n",
    enforce_stationarity=False,
    enforce_invertibility=False
)

# Train the best SARIMAX model with higher iteration limit for better convergence --
best_sarimax_fit = best_sarimax_model.fit(
    disp=False,
    maxiter=200
)

print(best_sarimax_fit.summary())
```

### 4.3.9 Future Forecasting

```python
# Initialize final SARIMAX model using the full dataset --
final_sarimax_model = SARIMAX(
    y_ts,
    exog=exog_ts,
    order=best_order,
    seasonal_order=best_seasonal_order,
    trend="n",
    enforce_stationarity=False,
    enforce_invertibility=False
)

# Train final SARIMAX model on all available data --
final_sarimax_fit = final_sarimax_model.fit(
    disp=False,
    maxiter=200
)

# Define forecast horizon (next 14 days) --
future_steps = 14

# Generate future dates starting after the last available observation --
future_dates = pd.date_range(
    start=y_ts.index[-1] + pd.Timedelta(days=1),
    periods=future_steps,
    freq="D"
)

# Define function to assign Philippine season based on future date --
def assign_ph_season_by_date(date):
    month = date.month

    if month in [11, 12, 1, 2]:
        return "Cool-Dry"
    elif month in [3, 4, 5]:
        return "Hot-Dry"
    else:
        return "Rainy"

# Create future seasonal labels for forecast dates --
future_seasons = pd.Series(
    [assign_ph_season_by_date(date) for date in future_dates],
    index=future_dates,
    name="season"
)

# Convert future seasonal labels into dummy variables --
future_exog = pd.get_dummies(
    future_seasons,
    prefix="season",
    dtype=int
)

# Ensure forecast exogenous variables match training exogenous columns --
future_exog = future_exog.reindex(
    columns=sarimax_exog_cols,
    fill_value=0
)

# Generate future forecast using trained final SARIMAX model --
future_forecast = final_sarimax_fit.get_forecast(
    steps=future_steps,
    exog=future_exog
)

# Extract forecasted mean values and confidence intervals --
future_mean = future_forecast.predicted_mean
future_ci = future_forecast.conf_int()

# Create dataframe containing forecast results --
future_results = pd.DataFrame({
    "date": future_dates,
    "forecasted_heat_index": future_mean.values,
    "lower_ci": future_ci.iloc[:, 0].values,
    "upper_ci": future_ci.iloc[:, 1].values
})

# Classify forecasted heat risk levels --
future_results["forecasted_heat_risk"] = future_results[
    "forecasted_heat_index"
].apply(classify_heat_risk)

display(future_results)
```
